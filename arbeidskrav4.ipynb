```python
# importer pakkene som vi trenger
import numpy as np
from matplotlib import pyplot as plt
import sympy as sp
```

```python
# definer symboler
N, A = sp.symbols('N A', positive=True, real=True)
```

```python
# Vi definerer produktfn

def prod(c,N,A):
    produksjon=1000*(c.exp(N)/(A+c.exp(N)))
    return produksjon
prod(sp,N,A)
```

```python
# lager funksjon
fun1=sp.diff(prod(sp,N,A),N)
fun1
```

```python
# bruker lamdify som sagt i oppgaven
fun1=sp.lambdify(
    (A,N),fun1)
```

```python
def fun2(N,A):
    return 1000*(np.exp(N)/(A+np.exp(N)))/N
```

```python
n=np.linspace(1,30,100)

# setter spesifikasjoner og navn
fig1, ax = plt.subplots()
ax.set_ylabel('Kvantum (x)')
ax.set_xlabel('Arbeidskraft (N)')
ax.set(xlim=(1,10))
ax.set(ylim=(0,300))
                
# lager grafene
ax.plot(n, fun1(200,n), label='Grenseproduktivitet')
ax.plot(n, fun2(n,200), label='Gjennomsnittlig produktivitet')

# Tittel
ax.set_title('Produksjon')

# legende
ax.legend();
```
